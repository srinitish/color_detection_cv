```python
from PIL import Image
from util import rgb_to_hsv_range
import cv2

vedio = cv2.VideoCapture(0)
blue = [255,0,0]
green = [0,0,255]
while True:
    ret,frame = vedio.read()
    hsv_img = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    
    lower,upper = rgb_to_hsv_range(blue)
    lower_g,upper_g = rgb_to_hsv_range(green)

    mask = cv2.inRange(hsv_img,lower,upper)
    mask_g = cv2.inRange(hsv_img,lower_g,upper_g)
    
    mask_g_ = Image.fromarray(mask_g)
    mask_ = Image.fromarray(mask)
   
    boundary_g  = mask_g_.getbbox()
    boundary = mask_.getbbox()

    print(boundary_g)
    if boundary_g is not None:
        xg1,yg1,xg2,yg2 = boundary_g
        frame =cv2.rectangle(frame,(xg1,yg1),(xg2,yg2),(0,255,0),3)
    
    if boundary is not None:
        x1,y1,x2,y2 = boundary
        frame =cv2.rectangle(frame,(x1,y1),(x2,y2),(255,0,0),3)
    

    cv2.imshow('mask',frame)
    if cv2.waitKey(1)& 0xff == ord(' '):
        break
vedio.release()
cv2.destroyAllWindows()

```
